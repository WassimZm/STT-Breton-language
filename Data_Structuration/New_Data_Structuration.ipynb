```python
import pandas as pd
import re
```

```python

with open('/home/ouassim/Downloads/350617.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# Prepare a list to hold each parsed line
data = []

for line in lines:
    # Use regex to split based on 2 or more whitespace characters (spaces or tabs)
    parts = re.split(r'\s{2,}|\t+', line.strip())
    
    # Ensure that the line has at least 8 parts (7 fields + text)
    if len(parts) >= 8:
        id_ = parts[0]
        start_time = parts[1]
        start_secs = parts[2]
        end_time = parts[3]
        end_secs = parts[4]
        duration_time = parts[5]
        duration_secs = parts[6]
        text = ' '.join(parts[7:])  # in case text has extra spaces
        
        data.append([
            id_, start_time, float(start_secs),
            end_time, float(end_secs),
            duration_time, float(duration_secs),
            text
        ])

```

```python
# Create DataFrame
df = pd.DataFrame(data, columns=[
    'id', 'start_time', 'start_secs',
    'end_time', 'end_secs',
    'duration_time', 'duration_secs',
    'text'
])
```

```python
import os
from pydub import AudioSegment

# Load the full audio
full_audio = AudioSegment.from_wav("/home/ouassim/Desktop/Stage-Breton/Data/Yuna Data/350617.wav")

# Define output folder
output_folder = "/home/ouassim/Desktop/Stage-Breton/Data/Yuna Data/Splitted Data"

# Create the folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Iterate over the DataFrame rows
for index, row in df.iterrows():
    start_ms = int(row['start_secs'] * 1000)
    end_ms = int(row['end_secs'] * 1000)
    
    # Slice the audio
    audio_clip = full_audio[start_ms:end_ms]
    
    # Create a file name based on the 'id' column
    filename = f"clip_{row['id_audio']}.wav"
    
    # Set the full path to save
    full_path = os.path.join(output_folder, filename)
    
    # Export the audio clip
    audio_clip.export(full_path, format="wav")
    print(f"Saved: {full_path}")

```
