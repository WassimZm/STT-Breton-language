```python
ground_truth_path="/kaggle/input/test-tsv/test.tsv"

ground_truth_dict = load_ground_truth_dict_commonvoice(ground_truth_path)
```

```python
from filter_char import filter_out_chars
from normalizer import normalize_sentence
from utils import pre_process


def process_br_text(br):
    PUNCTUATION = '<>.?!,;:«»“”"()[]/…–—•'
    br = filter_out_chars(br, PUNCTUATION + '*')
    br = normalize_sentence(br, autocorrect=True)
    br = pre_process(br).replace('-', ' ').lower()
    return br
```

```python
df['sentence_processed'] = df['sentence'].apply(process_br_text)
df['transcription_processed'] = df['transcription'].apply(process_br_text)
df['transcription_new_model_processed'] = df['transcription_new_model'].apply(process_br_text)
```

```python
df
```

```python
from jiwer import wer, cer

# Compute WER and CER for each row
df['wer_old_model'] = df.apply(lambda row: wer(row['sentence_processed'], row['transcription_processed']), axis=1)
df['cer_old_model'] = df.apply(lambda row: cer(row['sentence_processed'], row['transcription_processed']), axis=1)
```

```python
from jiwer import wer, cer

# Compute WER and CER for each row
df['wer_new_model'] = df.apply(lambda row: wer(row['sentence_processed'], row['transcription_new_model_processed']), axis=1)
df['cer_new_model'] = df.apply(lambda row: cer(row['sentence_processed'], row['transcription_new_model_processed']), axis=1)
```

```python
df = df[['sentence', 'sentence_processed', 'transcription_processed','transcription_new_model_processed', 'wer_new_model', 'wer_old_model', 'cer_new_model', 'cer_old_model']]
df
```

```python

average_wer_old = df['wer_old_model'].mean()
average_cer_old = df['cer_old_model'].mean()

# Optional: round them
average_wer_old = round(average_wer_old, 2)
average_cer_old = round(average_cer_old, 2)

print(f"Average old model WER: {average_wer_old}")
print(f"Average old model CER: {average_cer_old}")
```

```python

average_wer_new = df['wer_new_model'].mean()
average_cer_new = df['cer_new_model'].mean()

# Optional: round them
average_wer_new = round(average_wer_new, 2)
average_cer_new = round(average_cer_new, 2)

print(f"Average new model WER: {average_wer_new}")
print(f"Average new model CER: {average_cer_new}")
```
